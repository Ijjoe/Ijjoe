```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
import pyperclip
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os


options = Options()
options.add_experimental_option("prefs", {
    "profile.default_content_settings": {"javascript": 2}
})
#chrome_options = options
driver = webdriver.Chrome(options=options)
driver.get('https://kissjav.com/videos/jav-uncensored/9/')

try:
    time.sleep(10)
    # 웹 페이지의 모든 요소를 찾기
    print("ss")
    elements = driver.find_elements(By.XPATH,"//*")
    print("end")
    # 각 요소의 클래스 속성값을 출력
    for element in elements:
        class_name = element.get_attribute("class")
        if class_name:  # 클래스 속성이 비어있지 않은 경우에만 출력
            print(class_name)

 
except NoSuchElementException:
    print("Element not found")
finally:

    dr=driver.find_element(By.XPATH,'/html/body/div[4]/div[2]/div[2]/nav[2]/ul/li[1]/a')
    print(dr.text)

    time.sleep(300)
    driver.quit()

print('all ok')
```
